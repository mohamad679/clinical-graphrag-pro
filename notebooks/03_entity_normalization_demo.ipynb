# 03: Entity Normalization Demo

## Section 1 — Introduction

**UMLS** is a biomedical metathesaurus that links concepts across multiple medical vocabularies. It is useful as a normalization hub because one surface phrase can be mapped to a shared concept identifier and then traced across downstream code systems.

**SNOMED CT** is a comprehensive clinical terminology used to encode problems, findings, procedures, and other concepts in a granular way. It is well suited to semantic normalization because it captures clinical meaning rather than just billing categories.

**RxNorm** standardizes medication names across branded products, generic ingredients, and dose forms. It is especially useful in medication reconciliation and interaction workflows because different spellings and product names can be collapsed to the same normalized drug identity.

**ICD-10** is primarily a diagnostic classification system used widely for reporting and billing. It is less expressive than SNOMED CT for fine-grained semantics, but it remains important because many clinical and administrative workflows still revolve around ICD-coded diagnoses.

## Section 3 — Entity Type Distribution

This cell visualizes a simple distribution of normalized entity categories in a clinical extraction workload.

```python
labels = ['Conditions', 'Medications', 'Labs', 'Procedures']
sizes = [40, 35, 15, 10]
colors = sns.color_palette('muted', n_colors=len(labels))

fig, ax = plt.subplots(figsize=(7, 7))
ax.pie(sizes, labels=labels, autopct='%1.0f%%', startangle=90, colors=colors, wedgeprops={'linewidth': 1, 'edgecolor': 'white'})
ax.set_title('Illustrative Entity Type Distribution')
plt.show()
```

## Section 4 — Normalization Pipeline Diagram

This cell uses matplotlib patches to draw a simple conceptual pipeline from raw clinical text to a normalized entity.

```python
fig, ax = plt.subplots(figsize=(14, 3))
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')

steps = [
    {'label': 'Raw Text', 'x': 0.02, 'color': '#dbeafe'},
    {'label': 'SciSpaCy NER', 'x': 0.22, 'color': '#ede9fe'},
    {'label': 'UMLS Lookup', 'x': 0.42, 'color': '#dcfce7'},
    {'label': 'Fallback LLM', 'x': 0.62, 'color': '#fee2e2'},
    {'label': 'Normalized Entity', 'x': 0.82, 'color': '#fef3c7'},
]

for step in steps:
    box = mpatches.FancyBboxPatch(
        (step['x'], 0.38),
        0.15,
        0.22,
        boxstyle='round,pad=0.02',
        facecolor=step['color'],
        edgecolor='#334155',
        linewidth=1.5,
    )
    ax.add_patch(box)
    ax.text(step['x'] + 0.075, 0.49, step['label'], ha='center', va='center', fontsize=11, fontweight='bold')

for left, right in zip(steps[:-1], steps[1:]):
    ax.annotate(
        '',
        xy=(right['x'], 0.49),
        xytext=(left['x'] + 0.15, 0.49),
        arrowprops=dict(arrowstyle='->', lw=1.8, color='#475569'),
    )

ax.set_title('Entity Normalization Pipeline')
plt.tight_layout()
plt.show()
```
